# Variabel

Variabel merupakan "sesuatu" yang digunakan untuk menyimpan nilai. Variabel juga dapat digunakan sebagai referensi atau rujukan terhadap 'suatu' nilai. Variabel sangat berguna agar kita tidak perlu menulis ulang kode yang sama berulang-kali, karena kode telah disimpan pada variabel dan dapat dipanggil kapanpun dibutuhkan. Selain itu, variabel dapat ditulis bebas serta dapat diubah nilainya kapapun kita mau. Jika diibaratkan, variabel itu seperti container yang memuat barang-barang. Container disebut variabel, sedangkan muatan barang disebut nilai atau value.

Perlu diperhatikan dalam membuat suatu variabel:
- Penulisan variabel dimulai dari huruf (A - Z)/(Z-A), tidak boleh dimulai dengan angka
- Penulisan variabel bisa dimulai dengan huruf kecil maupun huruf besar
- Penulisan variabel bisa juga ditambahkan (_) contoh nama_lengkap
- Variabel bersifat case sensitive, artinya jika kita membuat dengan huruf kecil, kita tidak dapat memanggil variabel yang kita buat tersebut dengan huruf besar, harus huruf kecil juga. Contoh kita buat variabel nama = 'Hadi' kita tidak bisa memanggil variabel tersebut dengan NAMA, harus nama.

format membuat variabel:
nama_variabel = nilai

Dalam Python simbol "=" (sama dengan) diartikan sebagai penugasan (assignment) bukan kesetaraan (equality)

# Membuat Variabel

Kita akan mempraktekan membuat variabel dengan soal terkait investasi di materi sebelumnya

Soal: Kamu memiliki uang sebesar 300 ribu rupiah, yang mana akan diinvestasikan dengan return sebesar 10%. Berapa banyak uang kamu jika diinvestasikan selama 5 tahun.   

```python
# membuat variabel tabungan 300 ribu rupiah
tabungan = 300000
```

```python
# Cara memanggil variabel tabungan yang telah kita buat
print(tabungan)
```

```python
# Membuat variabel pertumbuhan investasi 10 % atau 1.1
pertumbuhan_investasi = 1.1

print(pertumbuhan_investasi)
```

```python
# Membuat variabel tahun atau 5
tahun = 5

print(tahun)
```

```python
# memanggil variabel yang telah dibuat dan kemudian dilakukan perhitungan seperti pada soal sebelumnya terkait investasi.
hasil_investasi = tabungan * pertumbuhan_investasi ** tahun

print(hasil_investasi)
```

# Membuat variabel dengan tipe data lain

```python
nama_depan = 'Muhamad'
nama_belakang = 'Hadiyansyah'
nama_lengkap = nama_depan + " " + nama_belakang
print(nama_lengkap)
```

```python
umur = 21
hobi = 'Berkuda'
print('Biodata\n', nama_lengkap, '\n', umur, '\n', hobi)
```

catatan: tanda petik dua ' ' itu menandakan nilai atau value dengan tipe data string, sedangkan tanpa petik dua ' ' itu menandakan varibel. '\n' itu berfungsi sebagai ENTER jika di microsoft word. Penggabungan variabel dapat menggunakan dua cara. 1. Dengan operasi perhitungan "+"  2. Dengan koma ","

```python
judul = 'SENJA DI JAKARTA'
penulis = 'Mochtar Lubis'
penerbit = 'BUKU OBOR'
tahun_terbit = 2019
print("Buku dari Dhira:\n", judul, '\n', penulis, '\n', penerbit, '\n', tahun_terbit)
```

```python
# Membuat variabel untuk menghitung luas persegi panjang
panjang = 10
lebar = 5
luas = panjang * lebar
print(luas)
```

```python
# Menghitungan PDB

# angka hanya perumpamaan
pdb_2018 = 4502

pdb_2019 = 5038

pdb = pdb_2019 - pdb_2018 / pdb_2018 * 100

print("PDB Indonesia Tahun 2019:", pdb) 
# Jika kita ingin menambahkan suatu kalimat pada saat menampilkan outpu, dapat memasukan suatu text diapit tanda kutip dua
```

```python
# Membuat biodata sederhana
nama = 'Kiwar'
usia = 23
sudah_menikah = False
print('nama: ', nama)
print('usia: ', usia)
print('sudah menikah: ', sudah_menikah)
```
